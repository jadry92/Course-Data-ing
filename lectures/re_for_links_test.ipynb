```python
import requests # library for request to the web page
import bs4 # library for manipulate web pages
import re # library for use regular Expressions 

try:
    response = requests.get('https://www.eltiempo.com')
#response? # ->this command give me the type of variable
except:
    print('Error')
    
#Extration of the html by bs4
soup = bs4.BeautifulSoup(response.text, 'html.parser')

# http://www.eltiempo.com/noticias/huella-social-12321
```

```python
def filter_links(regular_ex, links):
    links_valids = set([])
    for link in links:
        flag = regular_ex.match(link)
        if flag:
            links_valids.add(link)
    return list(links_valids)
```

```python
is_well_formed_link = re.compile(r'^https?://.+$')
is_root_path = re.compile(r'^/.+$')


def get_links_sub_pages(url):
    try:
        #print('access to ', url)
        response = requests.get(url)
        soup = bs4.BeautifulSoup(response.text, 'html.parser')
        links_news = set([])
        for link in soup.find_all('a'):
            link_full = link.get('href')
            if link_full:
                links_news.add(link_full)
        return list(links_news)
    except requests.exceptions.RequestException as e:  # This is the correct syntax
        print(e)
        return None
    

def sort_links(links):
    links_news = set([])
    for link in links:
        if is_well_formed_link.match(link):
            links_news.add(link)
        elif is_root_path.match(link):
            links_news.add('https://www.eltiempo.com{url}'.format(url = link))
    return list(links_news)
```

```python
##
pattern_main = re.compile(r'^https?://www.eltiempo.com.*$')
pattern_sub = re.compile(r'^https?://www.eltiempo.com.*[a-z]+$')
pattern_articles = re.compile(r'^https?://www.eltiempo.com.*-\d+$')

##
links_news = get_links_sub_pages('https://www.eltiempo.com')
links_news = sort_links(links_news)
links_news = filter_links(pattern_main, links_news)

##
sub_links = filter_links(pattern_sub, links_news)

##
links_articles = filter_links(pattern_articles, links_news)
print(len(links_news))
print(len(links_articles))
print(len(sub_links))

##
```

```python
links_visted = []

def access_to_sub_links(sub_links):
    global links_visted
    article_links = set([])
    for link in sub_links:
        if not link in links_visted:
            links_finded = get_links_sub_pages(link)
            links_visted.append(link)
            links_filtered_and_sorted = filter_links(pattern_main, sort_links(links_finded))
            article_links_list = filter_links(pattern_articles, links_filtered_and_sorted)
            article_links = article_links.union(set(article_links_list))
            sub_links += filter_links(pattern_sub, links_filtered_and_sorted)
            print('len = {}'.format(len(article_links)))
        if len(article_links) > 10000:
            return list(article_links)
    return list(article_links)

article_links = access_to_sub_links(sub_links)
print(len(article_links))
```
