# Equation of the Line #
These notes show different modes to define a line equation. The more common equation that we know $f(t)=m*t+b$. Where $m$ is the scope of the line and $b$ is the point where the line crosse the $f$ axis. however, another way to define a line is $f(t)=m*(t-a)$. Where $m$ is the scope of the line and $a$ is the point where the line crosse the $t$ axis.

It's possible to define the $b$ point in terms of the $a$ point if we take the last two-equation and applied the next process:

$$ f(t)=m*t+b \quad (1)$$
$$ f(t)=m*(t-a) \quad (2)$$
$$(1) - (2)$$
$$ (m*t+b) - (m*(t-a)) = f(t)- f(t) $$
$$ m*t+b - m*t+m*a = 0 $$
$$ b + m*a = 0 $$
$$ b = - m*a \quad (3)$$

And we can test this result with the next code:

```python
import matplotlib.pyplot as plt
import numpy as np

# time domnine
t = np.arange(-10,10,1)
# ecuation
m = 1 
a = 4
f = m*(t - a)
b = -a*m

fig, ax = plt.subplots()
ax.plot(t, f)
ax.plot(0,b,'or')
ax.plot(a,0,'ok')
ax.set(xlabel='time (s)', ylabel='$f$',
       title='$f(t)$')
ax.grid()
fig.show()
```
